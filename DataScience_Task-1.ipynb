```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter 
import plotly.express as px
```

```python
path = r"C:\Users\galig\Downloads\API_SP.POP.TOTL_DS2_en_excel_v2_311.xls"
df = pd.read_excel(path,skiprows=3)
```

```python
df.head()
```

```python
df_cleaned = df.dropna(subset=['Country Name'])
```

```python
df_long = df_cleaned.melt(id_vars=["Country Name"], 
                          var_name="Year", 
                          value_name="Population")
df_long = df_long.rename(columns={"Country Name": "Country"})
```

```python
df_long = df_long.dropna(subset=["Population"])
```

```python
df_long["Population"] = pd.to_numeric(df_long["Population"], errors='coerce')
```

```python
numeric_columns = df_long.select_dtypes(include='number').columns

top_countries = df_long.groupby('Country')[numeric_columns].mean().sort_values(by="Population", ascending=False).head(50).index

df_top_countries = df_long[df_long['Country'].isin(top_countries)]
```

```python
df_top_countries = df_top_countries.dropna(subset=["Population"])
```

```python
plt.figure(figsize=(15,10))
sns.set(style="whitegrid")
```

```python
fig = px.bar(
    df_top_countries,
    x="Population",
    y="Country",
    animation_frame="Year",
    title="Population of Top 30 Countries Across Years",
    labels={"Population": "Population", "Country": "Country"},
    orientation="h",
    color="Country",
    height=800,
    width=1000
)
```

```python
fig.update_layout(
    xaxis_title="Population",
    yaxis_title="Country",
    xaxis=dict(tickformat=",", title_font_size=18),
    yaxis=dict(title_font_size=18),
    title=dict(font_size=24),
    transition_duration=0
fig.update_traces(text=df_top_countries['Population'].apply(lambda x: f"{x:,.0f}"),
                  textposition='inside', textfont_size=15, insidetextanchor='start')
fig.show()
)
```
